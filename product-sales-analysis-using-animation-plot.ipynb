```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
```

```python
df.columns
```

```python
cols = ['Date','Profit', 'Margin', 'Sales', 'COGS', 'Total Expenses', 'Marketing', 'Inventory',
       'Budget Profit', 'Budget COGS', 'Budget Margin', 'Budget Sales']

subsetdf = df[cols]

subsetdf.set_index('Date',inplace=True)

cumsum_df = subsetdf.cumsum(axis=0)

cumsum_df
```

```python
#change to daily frequency
count_date = cumsum_df.groupby(cumsum_df.index)[['Profit', 'Margin', 'Sales', 'COGS', 'Total Expenses', 'Marketing', 'Inventory',
       'Budget Profit', 'Budget COGS', 'Budget Margin', 'Budget Sales']].sum()

pw_clean = pd.DataFrame(count_date)
pw_clean['Date'] = pd.to_datetime(pw_clean.index) 
pw_clean = pw_clean.set_index('Date')

pw_clean.head(20)
```
